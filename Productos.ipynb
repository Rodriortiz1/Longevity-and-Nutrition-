```python
import pandas as pd
from scipy.stats import pearsonr

df_azucar = df_unido_new[df_unido_new['Producto'] == 'Azucar y Dulcificantes']
df_bebidas = df_unido_new[df_unido_new['Producto'] == 'Bebidas Alcoholicas']
df_miscelaneo = df_unido_new[df_unido_new['Producto'] == 'Miscelaneo']

corr_azucar, p_value_azucar = pearsonr(df_azucar['Valor'], df_azucar['Tasa de obesidad'])
corr_bebidas, p_value_bebidas = pearsonr(df_bebidas['Valor'], df_bebidas['Tasa de obesidad'])
corr_miscelaneo, p_value_miscelaneo = pearsonr(df_miscelaneo['Valor'], df_miscelaneo['Tasa de obesidad'])

print(f"Correlación para Azúcar y Dulcificantes: {corr_azucar} (p-value: {p_value_azucar})")
print(f"Correlación para Bebidas Alcohólicas: {corr_bebidas} (p-value: {p_value_bebidas})")
print(f"Correlación para Misceláneo: {corr_miscelaneo} (p-value: {p_value_miscelaneo})")
```
