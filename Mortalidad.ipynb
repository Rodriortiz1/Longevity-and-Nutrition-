```python
import pandas as pd 
df_mortalidad = pd.read_csv("datos_paises4.csv")
df_mortalidad
```

```python
import re 
```

```python
def extraer_numero(cadena):
    # Usar una expresión regular para encontrar el número y el símbolo de porcentaje
    match = re.search(r'>([\d,]+‰)<', cadena)
    if match:
        return match.group(1)
    return None

# Aplicar la función a la lista de cadenas
df_mortalidad['2'] = [extraer_numero(cadena) for cadena in df_mortalidad["2"]]

# Mostrar los resultados
df_mortalidad
```

```python
df_mortalidad= df_mortalidad[df_mortalidad["1"].isin([(year) for year in range (2016,2022)])]
df_mortalidad
```

```python
def convertir_valor (valor):
    if pd.isnull(valor):
        return None
    valor_s = valor.replace("‰", "")
    valor_c = valor_s.replace(",",".")
    return float(valor_c)/1000
df_mortalidad["Tasa de mortalidad"] = df_mortalidad["Tasa de mortalidad"].apply(convertir_valor)

df_mortalidad
```
